### Basic Usage with ChatOpenAI and OpenRouter

Here's how to set up the `ChatOpenAI` class to connect to the Mistral Devstral 2512 Free model through OpenRouter.

````python
import os
from langchain_openai import ChatOpenAI
from google.colab import userdata

# It's recommended to set your API key as an environment variable
api_key = userdata.get("OPENROUTER_API_KEY")

# Initialize the ChatOpenAI model for Mistral Devstral 2512 Free
devstral_llm = ChatOpenAI(
    model="mistralai/devstral-2512:free",
    openai_api_key=api_key,
    openai_api_base="https://openrouter.ai/api/v1"
)

# Let's test it with a simple coding question
response = devstral_llm.invoke("What is recursion in programming and how does it work? Provide a simple example in Python.")
print(response.content)
````

### Building a Code Assistant Chatbot

We can create a conversational coding assistant by managing the chat history.

````python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="You are an expert programming assistant specialized in helping developers write clean, efficient, and maintainable code. You provide clear explanations and follow best practices."),
    HumanMessage(content="I need to implement a rate limiter in Python. Can you help me with that?"),
]

# First turn
response = devstral_llm.invoke(messages)
print(f"Code Assistant: {response.content}")

# Add the bot's response to the history
messages.append(response)

# Second turn
messages.append(HumanMessage(content="Great! Can you also add a decorator to apply this rate limiter to any function?"))
response = devstral_llm.invoke(messages)
print(f"\nCode Assistant: {response.content}")
````

## Advanced Examples

### Advanced Parameter Tuning

You can control Devstral 2512's output by tuning parameters like `temperature` and `max_tokens`.

- **`temperature`**: Controls creativity. Lower values (e.g., 0.1) make output more deterministic and focused, higher values (e.g., 0.9) allow more creative solutions.
- **`max_tokens`**: Sets the maximum length of the generated response.

````python
# Deterministic response with low temperature (best for production code)
deterministic_llm = ChatOpenAI(
    model="mistralai/devstral-2512:free",
    openai_api_key=api_key,
    openai_api_base="https://openrouter.ai/api/v1",
    temperature=0.1
)

prompt = "Write a Python function to validate an email address using regex. Make it production-ready with proper error handling."
deterministic_response = deterministic_llm.invoke(prompt)
print(f"Deterministic Code (temp=0.1):\n{deterministic_response.content}")

print("\n" + "="*60 + "\n")

# Creative response with higher temperature (for exploring alternatives)
creative_llm = ChatOpenAI(
    model="mistralai/devstral-2512:free",
    openai_api_key=api_key,
    openai_api_base="https://openrouter.ai/api/v1",
    temperature=0.7
)

prompt = "Suggest 3 creative ways to implement a caching mechanism in a web application."
creative_response = creative_llm.invoke(prompt)
print(f"Creative Suggestions (temp=0.7):\n{creative_response.content}")
````

### Streaming Responses

For longer code generations, you can use streaming to get real-time output.

````python
import time

# Create a streaming version of the model
streaming_llm = ChatOpenAI(
    model="mistralai/devstral-2512:free",
    openai_api_key=api_key,
    openai_api_base="https://openrouter.ai/api/v1",
    streaming=True
)

prompt = "Create a complete REST API using FastAPI with CRUD operations for a User model, including authentication with JWT tokens and proper error handling."

print("Streaming response:")
print("-" * 60)

# Stream the response
for chunk in streaming_llm.stream(prompt):
    print(chunk.content, end="", flush=True)
    time.sleep(0.01)  # Small delay to visualize streaming

print("\n" + "-" * 60)
print("âœ… Streaming complete!")
````

### Code Refactoring and Optimization

### Unit Test Generation

Let's test Devstral 2512's ability to generate comprehensive unit tests.

````python
# Function to test
function_code = """
def calculate_discount(price: float, discount_percentage: float, is_member: bool) -> float:
    if price < 0 or discount_percentage < 0 or discount_percentage > 100:
        raise ValueError("Invalid input")

    discount = price * (discount_percentage / 100)
    final_price = price - discount

    if is_member:
        final_price *= 0.95  # Additional 5% off for members

    return round(final_price, 2)
"""

test_prompt = f"""Generate comprehensive pytest unit tests for this function. Include:
1. Normal cases
2. Edge cases
3. Error cases
4. Parametrized tests where appropriate

{function_code}
"""

test_response = devstral_llm.invoke(test_prompt)
print(test_response.content)
````

### Performance Comparison

Let's compare Devstral 2512 with different temperature settings for various task types.

````python
import time

def test_model_performance(prompt, temperatures=[0.1, 0.5, 0.9]):
    """Test the same prompt with different temperature settings"""
    results = {}

    for temp in temperatures:
        print(f"\nðŸŒ¡ï¸ Testing with temperature: {temp}")
        print("-" * 50)

        test_llm = ChatOpenAI(
            model="mistralai/devstral-2512:free",
            openai_api_key=api_key,
            openai_api_base="https://openrouter.ai/api/v1",
            temperature=temp,
            max_tokens=300
        )

        start_time = time.time()
        response = test_llm.invoke(prompt)
        end_time = time.time()

        results[temp] = {
            'response': response.content,
            'time': end_time - start_time
        }

        print(f"Response: {response.content[:200]}...")  # Show first 200 chars
        print(f"â±ï¸ Time taken: {end_time - start_time:.2f} seconds")

    return results

# Test with algorithm implementation
algo_prompt = "Write a Python function to implement binary search with detailed comments."
print("ðŸ” DETERMINISTIC TASK: Binary search implementation")
print("=" * 60)

algo_results = test_model_performance(algo_prompt)

print("\n\nðŸ’¡ CREATIVE TASK: Suggesting architectural patterns")
print("=" * 60)

# Test with architectural question
arch_prompt = "Suggest creative ways to implement authentication in a distributed microservices system."
arch_results = test_model_performance(arch_prompt)
````
