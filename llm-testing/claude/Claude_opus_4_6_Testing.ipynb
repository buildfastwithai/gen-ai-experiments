## 3. Model Initialization
Initialize the OpenAI client pointing to OpenRouter.

```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)

MODEL_NAME = "anthropic/claude-opus-4.6"
```

## 4. Basic Testing
Let's verify the connection with a simple "Hello World" implementation.

```python
def get_completion(prompt):
    try:
        completion = client.chat.completions.create(
            extra_headers={
                "HTTP-Referer": "https://colab.research.google.com", # Optional, for OpenRouter rankings
                "X-Title": "Claude Opus Colab" # Optional
            },
            model=MODEL_NAME,
            messages=[
                {"role": "user", "content": prompt},
            ],
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"Error: {e}"

print(get_completion("Hello! Introduce yourself briefly."))
```

## 5. Capabilities (Simple)
Demonstrating core competencies: Text Generation and Code Generation.

```python
# Text Generation Task
prompt_text = "Explain the concept of 'Agentic Workflow' in 3 simple bullet points."
print("--- Text Generation ---")
print(get_completion(prompt_text))

# Code Generation Task
prompt_code = "Write a Python function to check for palindromes. Include comments."
print("\n--- Code Generation ---")
print(get_completion(prompt_code))
```

## 6. Capabilities (Advanced)
Exploring complex reasoning. Claude Opus is known for its ability to follow complex instructions.

```python
complex_prompt = """
Analyze the following sentence: "The quick brown fox jumps over the lazy dog."
1. Identify the part of speech for each word.
2. Explain why this sentence is famous.
3. Rewrite it in the style of Shakespeare.
Output in Markdown format.
"""
print(get_completion(complex_prompt))
```

## 7. Agentic Capabilities (with Agno)
Using **Agno** to create a simple agent that can search the web (simulated or using tools if configured) or just reason effectively.

*Note: For web search, you would typically add a tool like `Tavily` or `DuckDuckGo`. Here we demonstrate the basic Agent structure.*

```python
from agno.agent import Agent
from agno.models.openrouter import OpenRouter
from agno.tools.hackernews import HackerNewsTools

# Ensure the model supports tools if you add them.
# For this demo, we create a specialized 'Math Tutor' agent.


math_agent = Agent(
    model=OpenRouter(id=MODEL_NAME),
    tools=[HackerNewsTools()],
    add_history_to_context=True,
    description="You are a helpful and enthusiastic math tutor.",
    instructions=["Explain your steps clearly.", "Use emojis to make it fun."],
    markdown=True,
)

try:

    math_agent.print_response("If I have 3 apples and I eat one, but then I buy 5 more, how many do I have? Walk me through it!", stream=True)
except Exception as e:
    print(f"Agent creation failed (check model compatibility): {e}")
```
