```python
from langchain_openai import ChatOpenAI
from google.colab import userdata
from rich import print

llama3_405b = ChatOpenAI(model = "accounts/fireworks/models/llama-v3p1-405b-instruct",
                      openai_api_key = userdata.get("FIREWORKS_API_KEY"),
                      openai_api_base = "https://api.fireworks.ai/inference/v1"
)

print("Model initialized successfully!")
```

```python
## Logical Reasoning

response = llama3_405b.invoke("""
If 20 shirts take 10 hours to dry. How much time will it take to dry 100 shirts?
""")


print(response.content)
```
