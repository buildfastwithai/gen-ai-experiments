```python
# @title Configure OpenRouter Client
from google.colab import userdata
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get("OPENROUTER_API_KEY")
)

MODEL = "qwen/qwen3.5-35b-a3b"

print(f"✅ Client configured for: {MODEL}")
print(f"💡 35B total params | Only 3B active at inference time!")
```

## 💬 Example 1: Basic Chat & Reasoning

Test Qwen3.5-35B-A3B's fundamental chat ability and reasoning skills. With only 3B active params, let's see what it can do!

```python
# @title Basic Chat
response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": "Explain what Mixture-of-Experts (MoE) architecture is and why it makes AI inference more efficient. Use a simple analogy."}
    ],
    max_tokens=1024
)

print(response.choices[0].message.content)
print(f"\n📊 Tokens used: {response.usage.total_tokens}")
```

## 🧠 Example 2: Reasoning with Thinking Mode

Qwen3.5 supports **extended reasoning mode** — you can enable deep thinking for complex problems even with only 3B active parameters!

```python
# @title Reasoning Mode — Logic Puzzle
import json

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": """You have 12 identical-looking balls. One is heavier or lighter than the rest.
Using a balance scale only 3 times, find the odd ball and determine if it's heavier or lighter.
Walk through your reasoning step by step."""}
    ],
    max_tokens=2048,
    extra_body={
        "reasoning": {
            "effort": "high"
        }
    }
)

msg = response.choices[0].message

# Display reasoning if available
if hasattr(msg, 'reasoning_content') and msg.reasoning_content:
    print("💭 REASONING:")
    print("=" * 50)
    print(msg.reasoning_content)
    print("\n")

print("✅ ANSWER:")
print("=" * 50)
print(msg.content)
```

## 🛠️ Example 4: Tool Calling / Function Calling

Qwen3.5-35B-A3B is built for agentic workflows. Let's test its tool-calling capabilities with a multi-tool scenario.

```python
# @title Define Tools
import json

tools = [
    {
        "type": "function",
        "function": {
            "name": "search_web",
            "description": "Search the web for information on a topic",
            "parameters": {
                "type": "object",
                "properties": {
                    "query": {"type": "string", "description": "The search query"},
                    "max_results": {"type": "integer", "description": "Max number of results to return"}
                },
                "required": ["query"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "calculate",
            "description": "Evaluate a mathematical expression",
            "parameters": {
                "type": "object",
                "properties": {
                    "expression": {"type": "string", "description": "Math expression to evaluate (e.g. '2 ** 10 + 512')"}
                },
                "required": ["expression"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_model_info",
            "description": "Get technical information about an AI model",
            "parameters": {
                "type": "object",
                "properties": {
                    "model_name": {"type": "string", "description": "Name of the AI model"},
                    "fields": {
                        "type": "array",
                        "items": {"type": "string"},
                        "description": "Fields of info to fetch: params, architecture, context_length, price"
                    }
                },
                "required": ["model_name"]
            }
        }
    }
]

# Ask a question requiring multiple tools
response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": "Search for Qwen3.5-35B-A3B model info, calculate how many FLOPs saved vs a dense 35B model (35/3), and look up its architecture details."}
    ],
    tools=tools,
    tool_choice="auto",
    max_tokens=1024
)

message = response.choices[0].message

if message.tool_calls:
    print(f"🔧 Tool Calls Detected: {len(message.tool_calls)}\n")
    for i, tc in enumerate(message.tool_calls, 1):
        args = json.loads(tc.function.arguments)
        print(f"  Call {i}: {tc.function.name}")
        print(f"  Args: {json.dumps(args, indent=2)}")
        print()
else:
    print(message.content)
```

## 🌊 Example 5: Streaming — Real-time Output

Stream token-by-token responses for a better UX. With only 3B active params, streaming should be impressively fast!

```python
# @title Streaming Response
import time

print("📡 Streaming from Qwen3.5-35B-A3B:\n")

start = time.time()
first_token_time = None
char_count = 0

stream = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": "Write a beginner's guide to Mixture-of-Experts architecture in AI. Explain it so a 10-year-old can understand, using a classroom analogy."}
    ],
    max_tokens=1024,
    stream=True
)

full_response = ""
for chunk in stream:
    if chunk.choices[0].delta.content:
        content = chunk.choices[0].delta.content
        if first_token_time is None:
            first_token_time = time.time() - start
        full_response += content
        char_count += len(content)
        print(content, end="", flush=True)

total_time = time.time() - start
print(f"\n\n✅ Stream complete!")
print(f"⚡ Time to first token: {first_token_time:.2f}s")
print(f"⏱️  Total time: {total_time:.2f}s")
print(f"📝 Characters: {char_count}")
print(f"🚀 Speed: {char_count/total_time:.0f} chars/sec")
```

## 🏗️ Example 6: Structured JSON Output

Test JSON mode for reliable structured outputs — critical for production pipelines and data extraction.

```python
# @title Structured JSON Output
response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": """Respond ONLY in JSON with this exact schema:
{
  "model_analysis": {
    "name": string,
    "total_params": string,
    "active_params": string,
    "efficiency_ratio": string,
    "architecture": string,
    "best_use_cases": [string],
    "limitations": [string],
    "score": {"reasoning": number, "coding": number, "vision": number, "speed": number}
  }
}"""},
        {"role": "user", "content": "Analyze the Qwen3.5-35B-A3B model. Be specific and accurate."}
    ],
    response_format={"type": "json_object"},
    max_tokens=1024
)

result = json.loads(response.choices[0].message.content)
analysis = result["model_analysis"]

print("📊 Model Analysis — Qwen3.5-35B-A3B")
print("=" * 45)
print(f"📌 Name:              {analysis['name']}")
print(f"💾 Total Params:      {analysis['total_params']}")
print(f"⚡ Active Params:     {analysis['active_params']}")
print(f"📐 Efficiency Ratio:  {analysis['efficiency_ratio']}")
print(f"🏗️  Architecture:      {analysis['architecture']}")
print(f"\n✅ Best Use Cases:")
for uc in analysis["best_use_cases"]:
    print(f"   • {uc}")
print(f"\n⚠️  Limitations:")
for lim in analysis["limitations"]:
    print(f"   • {lim}")
print(f"\n🎯 Scores (out of 10):")
scores = analysis["score"]
for k, v in scores.items():
    bar = "█" * int(v) + "░" * (10 - int(v))
    print(f"   {k.capitalize():<12} [{bar}] {v}/10")
```

## 🤖 Example 7: Simple ReAct Agent Loop

Build a minimal agentic loop using Qwen3.5-35B-A3B's tool-calling capabilities.

```python
# @title Simple ReAct Agent

# Define simple agent tools
agent_tools = [
    {
        "type": "function",
        "function": {
            "name": "python_exec",
            "description": "Execute Python code and return result",
            "parameters": {
                "type": "object",
                "properties": {
                    "code": {"type": "string", "description": "Python code to execute"}
                },
                "required": ["code"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "write_file",
            "description": "Write content to a file",
            "parameters": {
                "type": "object",
                "properties": {
                    "filename": {"type": "string"},
                    "content": {"type": "string"}
                },
                "required": ["filename", "content"]
            }
        }
    }
]

def handle_tool(name, args):
    """Simulate tool execution"""
    if name == "python_exec":
        try:
            local_ns = {}
            exec(args["code"], local_ns)
            return f"✅ Code executed successfully. Output: {local_ns.get('result', 'No result variable set')}"
        except Exception as e:
            return f"❌ Error: {str(e)}"
    elif name == "write_file":
        return f"✅ File '{args['filename']}' written ({len(args['content'])} chars)"
    return "Tool executed"

# Run the agent
messages = [
    {"role": "system", "content": "You are a helpful coding agent. Use tools to complete tasks step by step."},
    {"role": "user", "content": "Write Python code to compute the first 10 Fibonacci numbers, execute it, then save the result to 'fibonacci.txt'."}
]

print("🤖 Starting ReAct Agent Loop\n")
print("=" * 50)

for step in range(6):
    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        tools=agent_tools,
        tool_choice="auto",
        max_tokens=2048
    )
    msg = response.choices[0].message
    messages.append(msg)

    if not msg.tool_calls:
        print(f"\n✅ Agent Done: {msg.content}")
        break

    for tc in msg.tool_calls:
        args = json.loads(tc.function.arguments)
        result = handle_tool(tc.function.name, args)
        print(f"Step {step+1}: 🔧 {tc.function.name}")
        print(f"   → {result}")
        messages.append({
            "role": "tool",
            "tool_call_id": tc.id,
            "content": result
        })
```

## ⚡ Example 8: Speed & Token Benchmarks

Measure exactly how fast Qwen3.5-35B-A3B is. With 3B active params, we expect excellent TTFT and throughput.

```python
# @title Benchmarks
import time

benchmark_tasks = [
    ("Simple Q&A",         "What is the capital of France?",                            256),
    ("Reasoning",          "If a snail travels at 0.03 mph, how long to cross 1 mile?", 512),
    ("Code Generation",    "Write a Python function to check if a number is prime.",    1024),
    ("Creative Writing",   "Write a haiku about neural networks.",                       256),
    ("Summarization",      "Summarize the MoE architecture in 3 bullet points.",        512),
]

results = []
print("⚡ Benchmarking Qwen3.5-35B-A3B")
print("=" * 65)
print(f"{'Task':<22} {'Latency':>10} {'Tokens':>8} {'Tok/s':>8}")
print("-" * 65)

for task_name, prompt, max_tok in benchmark_tasks:
    start = time.time()
    resp = client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tok
    )
    latency = time.time() - start
    tokens_out = resp.usage.completion_tokens
    tps = tokens_out / latency if latency > 0 else 0
    results.append((task_name, latency, tokens_out, tps))
    print(f"{task_name:<22} {latency:>9.2f}s {tokens_out:>8} {tps:>7.1f}")

print("-" * 65)
avg_latency = sum(r[1] for r in results) / len(results)
avg_tps = sum(r[3] for r in results) / len(results)
print(f"{'AVERAGE':<22} {avg_latency:>9.2f}s {'':>8} {avg_tps:>7.1f}")
print(f"\n💡 Pricing: https://openrouter.ai/qwen/qwen3.5-35b-a3b")
```
