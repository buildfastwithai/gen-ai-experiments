```python
from openai import OpenAI
from google.colab import userdata
from IPython.display import Image, display
import base64, requests, time

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get("OPENROUTER_API_KEY"),
)

MODEL = "google/gemini-3.1-flash-image-preview"

def show_images(response):
    """Display all images from an OpenRouter response."""
    msg = response.choices[0].message
    # Try response.images attribute
    if hasattr(msg, "images") and msg.images:
        for image in msg.images:
            image_url = image["image_url"]["url"]
            print(f"Image URL (first 80 chars): {image_url[:80]}...")
            display(Image(url=image_url))
    # Fallback: scan content list for image_url parts
    elif isinstance(msg.content, list):
        for part in msg.content:
            if hasattr(part, "type") and part.type == "image_url":
                url = part.image_url.url
                print(f"Image URL (first 80 chars): {url[:80]}...")
                display(Image(url=url))
    # Fallback: plain text (may contain URL)
    else:
        print("Response content:", msg.content)

print("Setup complete! Model:", MODEL)
```

## Example 1: Landscape Scene

Generate a beautiful natural landscape with dramatic lighting.

```python
response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {
            "role": "user",
            "content": "Generate a beautiful sunset over snow-capped mountains with golden hour lighting, a calm lake reflecting the colors, ultra-photorealistic, 8K"
        }
    ],
    extra_body={"modalities": ["image", "text"]}
)

print("Example 1: Landscape Scene")
show_images(response)
```

## Example 6: Image Editing

Pass an existing image and ask the model to edit it conversationally.

> First we generate a base image, then edit it in a follow-up call.

```python
# Step 1: Generate base image
base_response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {
            "role": "user",
            "content": "Generate a simple white ceramic coffee mug on a wooden table, soft daylight, photorealistic"
        }
    ],
    extra_body={"modalities": ["image", "text"]}
)

print("Base Image:")
show_images(base_response)

# Grab the image URL from the base response
base_msg = base_response.choices[0].message
base_image_url = None

if hasattr(base_msg, "images") and base_msg.images:
    base_image_url = base_msg.images[0]["image_url"]["url"]
elif isinstance(base_msg.content, list):
    for part in base_msg.content:
        if hasattr(part, "type") and part.type == "image_url":
            base_image_url = part.image_url.url
            break

if base_image_url:
    print("Base image captured. Proceeding to edit...")
else:
    print("Could not capture base image URL")
```

```python
# Step 2: Edit the base image
if base_image_url:
    edit_response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {"url": base_image_url}
                    },
                    {
                        "type": "text",
                        "text": "Add steam rising from the mug and place it in a cozy Parisian cafe with warm amber lighting in the background"
                    }
                ]
            }
        ],
        extra_body={"modalities": ["image", "text"]}
    )

    print("Edited Image:")
    show_images(edit_response)
else:
    print("Skipping edit - base image URL not available")
```

## Bonus: Aspect Ratio Control

Control the output aspect ratio using `image_config` in `extra_body`.

```python
# Supported ratios: "1:1", "16:9", "9:16", "4:3", "3:4", "4:1", "1:4"
RATIOS = ["1:1", "16:9", "9:16"]

for ratio in RATIOS:
    print(f"\nGenerating aspect ratio: {ratio}")
    r = client.chat.completions.create(
        model=MODEL,
        messages=[
            {
                "role": "user",
                "content": f"A beautiful cherry blossom tree in spring, Japan, soft pink petals falling, watercolor painting style"
            }
        ],
        extra_body={
            "modalities": ["image", "text"],
            "image_config": {"aspect_ratio": ratio}
        }
    )
    show_images(r)
```
