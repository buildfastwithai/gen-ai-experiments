# Rigorous Testing: google/gemini-3.1-pro-preview (via OpenRouter)

This notebook provides a comprehensive evaluation of **Google's Gemini 3.1 Pro Preview**, accessed through the OpenRouter API. We use **Agno** agents to test its advanced reasoning, extensive context window (up to 1M tokens), complex coding capabilities, and agentic workflows.

### 1. Setup & Installation
Install the required libraries: `agno`, `openai` (required for OpenRouter backend in Agno), and `duckduckgo-search` for realistic agentic tool usage.

```python
!pip install -q google-genai agno ddgs
```

### 2. Environment Variables & Model Initialization
Set your OpenRouter API key using Colab secrets. We'll instantiate the Gemini 3.1 Pro Preview model using Agno's OpenRouter wrapper.

```python
import os
from google.colab import userdata
from agno.models.google import Gemini
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")


from agno.agent import Agent
from agno.models.openrouter import OpenRouter

model_id = "gemini-3.1-pro-preview"

# Create an Agno Agent
agent = Agent(
    model=Gemini(id=model_id),
    description="You are an advanced AI assistant specialized in reasoning and coding.",
    markdown=True,
)
```

### 3. Basic Capability: Conversational Reasoning
Test the model's ability to handle complex logical puzzles, showcasing the "medium thinking level" capabilities of Gemini 3.1 Pro.

```python
prompt = """
Three friends (Alice, Bob, and Charlie) are playing a game.
One is a Knight (always tells the truth), one is a Knave (always lies), and one is a Spy (can do either).
Alice says: 'Bob is the Knave.'
Bob says: 'Charlie is the Spy.'
Charlie says: 'Alice is the Knight.'
Determine who is who, and walk me through your step-by-step reasoning.
"""

result = agent.run(prompt)
print(result.content)
```

### 4. Advanced Coding & Software Engineering Tasks
Gemini 3.1 Pro Preview boasts a 55.5 coding score. Let's ask it to write a complete Python application for a specific algorithmic problem.

```python
coding_prompt = """
Write a production-ready Python class that implements an LRU (Least Recently Used) Cache.
Requirements:
- Initialize with a given capacity.
- Include `get` and `put` methods operating in O(1) time complexity.
- Include docstrings and type hints.
- Provide a small test block using `assert`.
"""

result = agent.run(coding_prompt)
print(result.content)
```

### 5. Long Context Reading & Evaluation
Gemini 3.1 Pro has a 1M context window. While we won't pass 1 million tokens here, we will test its ability to process and summarize a dense block of text.

```python
long_text = """
Generative AI has undergone massive shifts over the last years. Models like GPT-4, Claude 3, and Gemini 1.5 pioneered the use of MoE (Mixture of Experts) to scale parameters without dramatically increasing compute costs.
Recently, the focus has shifted from parameter count to 'System 2' thinking and inference-time compute. Models like OpenAI's o1 and o3, and Google's Gemini 3 series, employ advanced reinforcement learning and internal chains-of-thought to deliberate before answering.
Agentic AI is the next frontier, moving from single-turn chatbots to autonomous workflows where models utilize tools, critique their own outputs, and adapt to changing environments. The Gemini 3.1 Pro Preview is explicitly optimized for these agentic loops, providing highly reliable JSON outputs and structured planning.
""" * 50  # Artificially expand the text to test scanning capability

context_prompt = f"""
Read the following text. Extract the three main evolutionary stages of Generative AI discussed, and format them as a JSON array of objects with keys 'stage' and 'description'.

<text>
{long_text}
</text>
"""

result = agent.run(context_prompt)
result.content
```

```python
from IPython.display import Markdown
Markdown(result.content)
```

### 6. Structured Output & Formatting
Testing exact formatting constraints, a critical skill for programmatic API usage.

```python
from pydantic import BaseModel, Field

class ModelReview(BaseModel):
    model_name: str
    strengths: list[str] = Field(..., max_items=3)
    score: int = Field(..., ge=1, le=10)

structured_agent = Agent(
    model=Gemini(id=model_id),
    output_schema=ModelReview
)

print("Generating Structured Output...")
response = structured_agent.run("Write a brief review of the Gemini 3.1 Pro Preview model based on our tests today.")
```

```python
print(response.content)
```
