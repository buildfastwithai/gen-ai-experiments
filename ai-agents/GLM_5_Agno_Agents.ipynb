## ğŸ” Example 1: Web Research Agent

Build an agent that can autonomously search the web using **Tavily** to answer research questions with up-to-date information.

**Use Case:** A research assistant that finds and synthesizes information from the web.

```python
# @title Web Research Agent
from agno.agent import Agent
from agno.tools.tavily import TavilyTools

research_agent = Agent(
    name="Web Researcher",
    model=glm5_model,
    tools=[TavilyTools()],
    description="You are an expert web researcher. Search the web to find accurate, up-to-date information and provide well-structured summaries with sources.",
    instructions=[
        "Always search the web before answering",
        "Cite your sources with URLs",
        "Provide a structured summary with key findings",
        "Highlight any conflicting information you find"
    ],
    markdown=True
)

research_agent.print_response(
    "What are the latest developments in open-source AI models in 2025? Focus on the top 3 most impactful releases.",
    stream=True
)
```

## ğŸ“ˆ Example 2: Financial Analyst Agent

Build an agent that can pull real-time stock data, analyze financials, and provide investment insights using **YFinance**.

**Use Case:** A financial analyst that helps track stocks and provides data-driven market analysis.

```python
# @title Financial Analyst Agent
from agno.tools.yfinance import YFinanceTools

finance_agent = Agent(
    name="Financial Analyst",
    model=glm5_model,
    tools=[YFinanceTools(
    )],
    description="You are a senior financial analyst. Analyze stocks using real market data and provide clear, actionable insights.",
    instructions=[
        "Always fetch real-time data before making analysis",
        "Present data in tables when possible",
        "Include both bull and bear perspectives",
        "Add a risk assessment to your analysis"
    ],
    markdown=True
)

finance_agent.print_response(
    "Compare NVDA and AMD stocks. Show current prices, analyst recommendations, and give me a brief investment outlook.",
    stream=True
)
```

---
## ğŸ“° Example 3: Tech News Summarizer Agent

Build an agent that fetches trending stories from **HackerNews** and provides concise, digestible summaries.

**Use Case:** A tech news curator that keeps you updated on what's trending in the developer community.

```python
# @title Tech News Summarizer
from agno.tools.hackernews import HackerNewsTools

news_agent = Agent(
    name="Tech News Curator",
    model=glm5_model,
    tools=[HackerNewsTools()],
    description="You are a tech news curator specializing in AI and software development trends. Summarize trending stories in an engaging way.",
    instructions=[
        "Fetch the top trending stories from HackerNews",
        "Summarize each story in 2-3 sentences",
        "Highlight why each story matters for developers",
        "Group stories by theme if possible (AI, DevTools, etc.)"
    ],
    markdown=True
)

news_agent.print_response(
    "What are the top 5 trending stories on HackerNews right now? Give me a quick summary of each.",
    stream=True
)
```

---
## ğŸ¢ Example 4: Multi-Agent Research Team

Combine all three agents into a **coordinated team** that delegates tasks automatically. The team leader routes each sub-question to the right specialist agent.

**Use Case:** A comprehensive research team that combines web research, financial analysis, and tech news into unified reports.

```python
# @title Multi-Agent Research Team
from agno.team import Team

# Reuse the agents defined above
research_team = Team(
    name="GLM-5 Research Team",
    members=[research_agent, finance_agent, news_agent],
    model=glm5_model,
    description="You are a research team coordinator. Delegate tasks to the most appropriate team member based on the query.",
    instructions=[
        "For web research questions, delegate to the Web Researcher",
        "For stock/financial questions, delegate to the Financial Analyst",
        "For tech news questions, delegate to the Tech News Curator",
        "Combine insights from multiple agents when the query spans multiple domains",
        "Provide a unified, well-structured final report"
    ],
    markdown=True
)

research_team.print_response(
    "I'm researching the AI chip market. Can you: 1) Find the latest news about AI chip companies from the web, 2) Compare NVDA vs AMD stock performance, and 3) Check what HackerNews is saying about AI hardware?",
    stream=True
)
```
