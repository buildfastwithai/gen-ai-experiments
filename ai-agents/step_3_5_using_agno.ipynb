## ğŸ¤– Example 1: Basic Agent with StepFun

We use `agno.agent.Agent` and configure it to use OpenRouter as the provider.

```python
from agno.agent import Agent
from agno.models.openrouter import OpenRouter

# Initialize StepFun Model via OpenRouter
stepfun_model = OpenRouter(
    id="stepfun/step-3.5-flash:free",
)

# Create simple agent
agent = Agent(
    model=stepfun_model,
    description="You are a helpful and enthusiastic AI assistant.",
    markdown=True
)

# Run
agent.print_response("Hello! What makes you special as a StepFun model?", stream=True)
```

## ğŸ” Example 2: Web Search Agent (Tavily)

We give the agent the `TavilyTools` to search the web autonomously.

```python
from agno.tools.tavily import TavilyTools

web_agent = Agent(
    model=stepfun_model,
    tools=[TavilyTools()],
    description="You are a researcher. Search the web for information if needed.",
    markdown=True
)

web_agent.print_response("What are the latest rumors about the iPhone 17 release date?", stream=True)
```

## ğŸ’¼ Example 3: Agent Team (Delegation)

We create a team of agents: a news and a finance agent.

```python
from agno.team import Team
from agno.agent import Agent
from agno.models.openai import OpenAIResponses
from agno.tools.hackernews import HackerNewsTools
from agno.tools.yfinance import YFinanceTools

news_agent = Agent(
    name="News Agent",
    description="Get trending tech news from HackerNews",
    model=stepfun_model,
    tools=[HackerNewsTools()]
)

finance_agent = Agent(
    name="Finance Agent",
    description="Get stock prices and financial data",
    model=stepfun_model,
    tools=[YFinanceTools()]
)

team = Team(
    name="Research Team",
    members=[news_agent, finance_agent],
    model=stepfun_model,
    instructions="Delegate to the appropriate agent based on the request."
)

team.print_response("What are the trending AI stories and how is NVDA stock doing?", stream=True)
```
