## ğŸ“ Example 1: Creating Your First Skill

A skill is a folder with a `SKILL.md` file containing YAML frontmatter and markdown instructions.

**Skill Structure:**
```
my-skill/
â”œâ”€â”€ SKILL.md          # Required: Instructions + YAML frontmatter
â”œâ”€â”€ scripts/          # Optional: Executable scripts
â”‚   â””â”€â”€ helper.py
â””â”€â”€ references/       # Optional: Reference documentation
    â””â”€â”€ guide.md
```

Let's create a **writing assistant** skill from scratch.

````python
# @title Create a Writing Assistant Skill
import os

# Create skill directory structure
SKILLS_DIR = "/content/skills"
os.makedirs(f"{SKILLS_DIR}/writing-assistant/references", exist_ok=True)

# 1. Create SKILL.md (the core instruction file)
skill_md = """---
name: writing-assistant
description: Helps write, edit, and improve text with professional writing guidance
license: Apache-2.0
metadata:
  version: "1.0.0"
  author: buildfastwithai
  tags: ["writing", "editing", "content"]
---

# Writing Assistant Skill

Use this skill when helping users write, edit, or improve text content.

## When to Use
- User asks for help writing emails, articles, or reports
- User wants to improve clarity, tone, or structure
- User needs proofreading or grammar corrections

## Process
1. **Understand Intent**: Clarify the purpose, audience, and tone
2. **Draft/Edit**: Write or revise the content
3. **Polish**: Check grammar, clarity, and flow
4. **Format**: Apply appropriate formatting (headers, bullets, etc.)

## Style Guidelines
- Use active voice over passive voice
- Keep sentences concise (aim for 15-20 words)
- Use specific, concrete language
- Break long paragraphs into shorter ones
- Start with the most important information

## Tone Profiles
- **Professional**: Formal, structured, objective
- **Casual**: Friendly, conversational, approachable
- **Technical**: Precise, detailed, terminology-rich
- **Persuasive**: Compelling, action-oriented, benefit-focused
"""

with open(f"{SKILLS_DIR}/writing-assistant/SKILL.md", "w") as f:
    f.write(skill_md)

# 2. Create a reference document
reference_md = """# Email Templates

## Follow-up Email
Subject: Following up on [topic]
Hi [Name], I wanted to follow up on our conversation about [topic]...

## Meeting Request
Subject: Meeting Request â€” [Topic]
Hi [Name], I'd like to schedule a meeting to discuss [topic]...

## Thank You Email
Subject: Thank you for [reason]
Hi [Name], I wanted to thank you for [reason]...
"""

with open(f"{SKILLS_DIR}/writing-assistant/references/email-templates.md", "w") as f:
    f.write(reference_md)

print("âœ… Writing Assistant skill created!")
print(f"ğŸ“ Location: {SKILLS_DIR}/writing-assistant/")

# Show the directory structure
for root, dirs, files in os.walk(f"{SKILLS_DIR}/writing-assistant"):
    level = root.replace(f"{SKILLS_DIR}/writing-assistant", "").count(os.sep)
    indent = " " * 2 * level
    print(f"{indent}ğŸ“‚ {os.path.basename(root)}/")
    subindent = " " * 2 * (level + 1)
    for file in files:
        print(f"{subindent}ğŸ“„ {file}")
````

- `get_skill_instructions(skill_name)` â€” Load full guidance

## ğŸ’» Example 3: Code Review Skill with Scripts & References

Build a comprehensive **code review skill** with:
- A SKILL.md with review process instructions
- A Python script for automated style checking
- A reference guide with coding standards

````python
# @title Create Code Review Skill (with Scripts + References)
import os

os.makedirs(f"{SKILLS_DIR}/code-review/scripts", exist_ok=True)
os.makedirs(f"{SKILLS_DIR}/code-review/references", exist_ok=True)

# 1. SKILL.md â€” Core instructions
code_review_skill = """---
name: code-review
description: Code review assistance with style checking, bug detection, and best practices
license: Apache-2.0
metadata:
  version: "1.0.0"
  author: buildfastwithai
  tags: ["python", "code-quality", "review"]
---

# Code Review Skill

Use this skill when reviewing code for quality, style, and best practices.

## When to Use
- User asks for code review or feedback
- User wants to improve code quality
- User needs help with refactoring
- User submits code with potential bugs

## Review Process
1. **Analyze Structure**: Review overall code organization
2. **Check Style**: Look for style guide violations (use the check_style.py script)
3. **Identify Issues**: Find bugs, security issues, performance problems
4. **Suggest Improvements**: Provide actionable recommendations with examples

## Severity Levels
- ğŸ”´ **Critical**: Bugs, security vulnerabilities, data loss risks
- ğŸŸ¡ **Warning**: Performance issues, code smells, missing error handling
- ğŸ”µ **Info**: Style suggestions, naming improvements, documentation

## Best Practices
- Focus on the most impactful issues first
- Explain the "why" behind every suggestion
- Provide code examples for each fix
- Be constructive, not critical
"""

with open(f"{SKILLS_DIR}/code-review/SKILL.md", "w") as f:
    f.write(code_review_skill)

# 2. Script â€” Automated style checker
check_style_script = '''#!/usr/bin/env python3
"""Check Python code style and return results."""
import sys

def check_style(code: str) -> dict:
    issues = []
    lines = code.split('\\n')

    for i, line in enumerate(lines, 1):
        # Line length check
        if len(line) > 100:
            issues.append(f"Line {i}: exceeds 100 characters ({len(line)} chars)")

        # Trailing whitespace
        if line.endswith(' '):
            issues.append(f"Line {i}: trailing whitespace")

        # Missing docstring detection (basic)
        stripped = line.strip()
        if stripped.startswith('def ') and i < len(lines):
            next_line = lines[i].strip() if i < len(lines) else ""
            if not next_line.startswith(('"""', "'\''\''")):
                issues.append(f"Line {i}: function missing docstring")

    return {"issues": issues, "count": len(issues), "lines_checked": len(lines)}

if __name__ == "__main__":
    code = sys.stdin.read() if not sys.argv[1:] else sys.argv[1]
    result = check_style(code)
    print(result)
'''

with open(f"{SKILLS_DIR}/code-review/scripts/check_style.py", "w") as f:
    f.write(check_style_script)

# 3. Reference â€” Python style guide
style_guide = """# Python Style Guide

## Naming Conventions
- **Variables**: `snake_case`
- **Classes**: `PascalCase`
- **Constants**: `UPPER_SNAKE_CASE`
- **Private**: `_leading_underscore`
- **Protected**: `__double_leading`

## Line Length
- Maximum 100 characters per line
- Break long lines at logical points

## Imports
- Standard library imports first
- Third-party imports second
- Local imports third
- Alphabetize within each group

## Type Hints
- Always add type hints to function signatures
- Use `Optional[T]` for nullable parameters
- Use `list[T]`, `dict[K, V]` (Python 3.9+)

## Docstrings
- Use Google-style docstrings
- Document all public functions and classes
- Include Args, Returns, and Raises sections
"""

with open(f"{SKILLS_DIR}/code-review/references/style-guide.md", "w") as f:
    f.write(style_guide)

print("âœ… Code Review skill created with scripts and references!")
print()

# Show structure
for root, dirs, files in os.walk(f"{SKILLS_DIR}/code-review"):
    level = root.replace(f"{SKILLS_DIR}/code-review", "").count(os.sep)
    indent = " " * 2 * level
    print(f"{indent}ğŸ“‚ {os.path.basename(root)}/")
    subindent = " " * 2 * (level + 1)
    for file in files:
        print(f"{subindent}ğŸ“„ {file}")
````

````python
# @title Create Code Review Agent and Test It
from agno.skills import Skills, LocalSkills

# Reload skills to pick up the new code-review skill
skills = Skills(loaders=[LocalSkills(SKILLS_DIR)])

code_review_agent = Agent(
    name="Code Reviewer",
    model=model,
    skills=skills,
    instructions=[
        "You are a senior code reviewer.",
        "Always load the code-review skill instructions before reviewing.",
        "Check the style-guide reference for coding standards.",
        "Categorize issues by severity: ğŸ”´ Critical, ğŸŸ¡ Warning, ğŸ”µ Info",
        "Provide fixed code examples for each issue."
    ],
    markdown=True
)

# Test with a code snippet that has several issues
code_review_agent.print_response(
    """Review this Python code for issues and improvements:

```python
import os,sys
import json
from typing import List

class userManager:
    def __init__(self, db_path):
        self.db = db_path
        self.users = []

    def get_user(self, id):
        for u in self.users:
            if u['id'] == id:
                return u
        return None

    def add_user(self, name, email, password):
        user = {'id': len(self.users)+1, 'name': name, 'email': email, 'password': password}
        self.users.append(user)

    def delete_all(self):
        self.users = []
```""",
    stream=True
)
````

## ğŸ“Š Example 4: Data Analysis Skill with Python Scripts

Create a **data analysis skill** that includes a reusable Python script for generating quick stats on datasets.

````python
# @title Create Data Analysis Skill
import os

os.makedirs(f"{SKILLS_DIR}/data-analysis/scripts", exist_ok=True)
os.makedirs(f"{SKILLS_DIR}/data-analysis/references", exist_ok=True)

# 1. SKILL.md
data_skill = """---
name: data-analysis
description: Data analysis assistance with statistical methods, visualization guidance, and insight generation
license: Apache-2.0
metadata:
  version: "1.0.0"
  author: buildfastwithai
  tags: ["data", "analytics", "statistics", "python"]
---

# Data Analysis Skill

Use this skill when analyzing data, generating insights, or creating visualizations.

## When to Use
- User provides a dataset or data description
- User asks for statistical analysis
- User wants help creating charts or dashboards
- User needs help interpreting data patterns

## Analysis Framework
1. **Understand the Data**: Check shape, types, missing values
2. **Descriptive Stats**: Mean, median, std, distributions
3. **Explore Patterns**: Correlations, trends, outliers
4. **Visualize**: Choose the right chart for the data type
5. **Insight**: Summarize findings in plain language

## Chart Selection Guide
- **Trend over time** â†’ Line chart
- **Comparison** â†’ Bar chart
- **Distribution** â†’ Histogram / Box plot
- **Relationship** â†’ Scatter plot
- **Composition** â†’ Pie chart / Stacked bar
- **Correlation** â†’ Heatmap

## Best Practices
- Always check for missing values first
- Use appropriate statistical tests
- Report confidence intervals, not just point estimates
- Visualize before modeling
"""

with open(f"{SKILLS_DIR}/data-analysis/SKILL.md", "w") as f:
    f.write(data_skill)

# 2. Script â€” Quick dataset profiler
profiler_script = '''#!/usr/bin/env python3
"""Quick dataset profiler â€” generates summary statistics."""
import json
import sys

def profile_data(data: list[dict]) -> dict:
    """Profile a list of dictionaries (rows) and return summary stats."""
    if not data:
        return {"error": "Empty dataset"}

    columns = list(data[0].keys())
    profile = {
        "rows": len(data),
        "columns": len(columns),
        "column_names": columns,
        "stats": {}
    }

    for col in columns:
        values = [row.get(col) for row in data if row.get(col) is not None]
        numeric = [v for v in values if isinstance(v, (int, float))]

        col_stats = {"non_null": len(values), "null": len(data) - len(values)}

        if numeric:
            col_stats.update({
                "type": "numeric",
                "min": min(numeric),
                "max": max(numeric),
                "mean": sum(numeric) / len(numeric),
            })
        else:
            unique = list(set(str(v) for v in values))
            col_stats.update({
                "type": "categorical",
                "unique_count": len(unique),
                "sample_values": unique[:5]
            })

        profile["stats"][col] = col_stats

    return profile

if __name__ == "__main__":
    data = json.loads(sys.stdin.read())
    result = profile_data(data)
    print(json.dumps(result, indent=2))
'''

with open(f"{SKILLS_DIR}/data-analysis/scripts/profiler.py", "w") as f:
    f.write(profiler_script)

# 3. Reference â€” Common statistical tests
stats_ref = """# Statistical Tests Reference

## When to Use Which Test

| Question | Test | Python |
|----------|------|--------|
| Are two means different? | t-test | `scipy.stats.ttest_ind()` |
| Is data normally distributed? | Shapiro-Wilk | `scipy.stats.shapiro()` |
| Are two variables correlated? | Pearson r | `scipy.stats.pearsonr()` |
| Compare 3+ groups? | ANOVA | `scipy.stats.f_oneway()` |
| Non-normal comparison? | Mann-Whitney U | `scipy.stats.mannwhitneyu()` |
"""

with open(f"{SKILLS_DIR}/data-analysis/references/statistical-tests.md", "w") as f:
    f.write(stats_ref)

print("âœ… Data Analysis skill created!")
print(f"   ğŸ“„ SKILL.md â€” Analysis framework + chart selection guide")
print(f"   ğŸ“œ scripts/profiler.py â€” Quick dataset profiler")
print(f"   ğŸ“– references/statistical-tests.md â€” Common tests reference")
````

````python
# @title Test the Data Analysis Agent
skills = Skills(loaders=[LocalSkills(SKILLS_DIR)])

data_agent = Agent(
    name="Data Analyst",
    model=model,
    skills=skills,
    instructions=[
        "You are a senior data analyst.",
        "Load the data-analysis skill for guidance on analysis methodology.",
        "Use the statistical-tests reference when recommending tests.",
        "Always suggest appropriate visualizations for the data."
    ],
    markdown=True
)

data_agent.print_response(
    """I have sales data for 4 regions over 12 months:
- North: avg $45K/mo, std dev $8K, trending up 3% monthly
- South: avg $38K/mo, std dev $12K, flat trend
- East: avg $52K/mo, std dev $6K, trending up 1% monthly
- West: avg $41K/mo, std dev $15K, trending down 2% monthly

How should I analyze this data? What statistical tests should I run? What visualizations would be most effective?""",
    stream=True
)
````

````python
# @title List All Loaded Skills
print("ğŸ“¦ All Skills in Directory:")
print("=" * 50)

for skill_dir in sorted(os.listdir(SKILLS_DIR)):
    skill_path = os.path.join(SKILLS_DIR, skill_dir, "SKILL.md")
    if os.path.exists(skill_path):
        with open(skill_path, 'r') as f:
            content = f.read()

        # Parse name and description from frontmatter
        name = desc = "N/A"
        if '---' in content:
            parts = content.split('---')
            if len(parts) >= 3:
                meta = parts[1]
                for line in meta.strip().split('\n'):
                    if line.startswith('name:'):
                        name = line.split(':', 1)[1].strip()
                    if line.startswith('description:'):
                        desc = line.split(':', 1)[1].strip()

        # Count components
        has_scripts = os.path.exists(os.path.join(SKILLS_DIR, skill_dir, 'scripts'))
        has_refs = os.path.exists(os.path.join(SKILLS_DIR, skill_dir, 'references'))

        components = ["SKILL.md"]
        if has_scripts: components.append("scripts/")
        if has_refs: components.append("references/")

        print(f"\nğŸ“¦ {name}")
        print(f"   ğŸ“ {desc}")
        print(f"   ğŸ“‚ Components: {', '.join(components)}")
````
