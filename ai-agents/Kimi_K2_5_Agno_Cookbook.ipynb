## ğŸ§  Agent 1: Deep Reasoning Agent

**Kimi K2.5 + Agno ReasoningTools = Unstoppable Problem Solver**

This agent combines Kimi K2.5's native reasoning with Agno's ReasoningTools for transparent, step-by-step thinking on complex problems.

```python
from agno.agent import Agent
from agno.models.openrouter import OpenRouter
from agno.tools.reasoning import ReasoningTools

print("ğŸ§  DEEP REASONING AGENT")
print("=" * 60)

# Create a reasoning-enhanced agent with Kimi K2.5
reasoning_agent = Agent(
    model=OpenRouter(id=KIMI_K25),
    tools=[ReasoningTools(add_instructions=True)],
    instructions="""You are a world-class problem solver. For complex problems:
    1. Use the think() tool to break down the problem
    2. Consider multiple approaches
    3. Validate your reasoning
    4. Provide a clear, structured answer""",
    markdown=True,
)

# Test with a complex reasoning problem
reasoning_agent.print_response(
    """A company has 100 employees. The CEO wants to form project teams with these rules:
    - Each team must have exactly 5 members
    - Each employee must be on at least 2 teams
    - No two teams can have more than 2 members in common

    Is this possible? If yes, how many teams are needed minimum? Show your reasoning.""",
    stream=True,
    show_full_reasoning=True,
)
```

## ğŸ¯ Agent 7: Production-Ready Template

**Deploy to AgentOS in 5 Minutes**

A complete, production-ready agent template using all of Agno's best practices.

```python
from textwrap import dedent
from typing import List, Optional
from pydantic import BaseModel, Field
from agno.agent import Agent
from agno.models.openrouter import OpenRouter, OpenRouterResponses
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.tools.reasoning import ReasoningTools
from agno.db.sqlite import SqliteDb

print("ğŸ¯ PRODUCTION-READY AGENT TEMPLATE")
print("=" * 60)

class TaskResult(BaseModel):
    """Structured output for any task."""
    success: bool = Field(..., description="Whether the task completed successfully")
    summary: str = Field(..., description="Brief summary of what was done")
    key_findings: List[str] = Field(..., description="Main findings or outputs")
    next_steps: Optional[List[str]] = Field(None, description="Recommended next actions")
    confidence_score: float = Field(..., description="0-1 confidence in the result")

class ProductionAgent:
    """
    Production-ready Kimi K2.5 agent with:
    - Fallback model routing
    - Persistent learning
    - Reasoning capabilities
    - Web search
    - Structured outputs
    """

    def __init__(
        self,
        name: str = "Production Agent",
        db_path: str = "tmp/production_agent.db",
        enable_learning: bool = True,
    ):
        self.agent = Agent(
            name=name,
            model=OpenRouterResponses(
                id="moonshotai/kimi-k2.5",
                models=[
                    "moonshotai/kimi-k2.5",
                    "google/gemini-2.0-flash-001",
                    "openai/gpt-4o-mini",
                ],
            ),
            tools=[
                DuckDuckGoTools(),
                ReasoningTools(add_instructions=True),
            ],
            db=SqliteDb(db_file=db_path) if enable_learning else None,
            learning=enable_learning,
            instructions=dedent("""\
                You are a highly capable AI assistant optimized for production use.

                Core behaviors:
                1. Always reason through complex problems using the think() tool
                2. Search the web when current information is needed
                3. Be thorough but efficient
                4. Admit uncertainty when appropriate
                5. Provide actionable, structured outputs

                Quality standards:
                - Accuracy over speed
                - Cite sources when using web information
                - Consider edge cases
                - Validate reasoning before responding"""),
            markdown=True,
            add_datetime_to_context=True,
        )

    def run(
        self,
        task: str,
        user_id: Optional[str] = None,
        session_id: Optional[str] = None,
        stream: bool = True,
    ):
        """Execute a task with the agent."""
        return self.agent.print_response(
            task,
            user_id=user_id,
            session_id=session_id,
            stream=stream,
        )

    def run_structured(
        self,
        task: str,
        user_id: Optional[str] = None,
    ) -> TaskResult:
        """Execute a task and return structured output."""
        structured_agent = Agent(
            model=self.agent.model,
            tools=self.agent.tools,
            response_model=TaskResult,
        )
        response = structured_agent.run(task)
        return response.content

# Create and test the production agent
agent = ProductionAgent(name="Kimi K2.5 Production Agent")

agent.run(
    "What are the key differences between Kimi K2.5 and other frontier models for agentic tasks? Provide a brief analysis.",
    stream=True,
)
```
